# 1.이미지에 있는 텍스트 추출 

```python
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
import sys, os
```

```python
IMAGE_PATH = 'D:\\data\hange.jpg'
img = cv2.imread(IMAGE_PATH)
plt.imshow(img)
```

```python
reader = easyocr.Reader(['ko', 'en'], gpu=False)
result = reader.readtext(IMAGE_PATH)
result
```

```python
sen=[]
for i in range(0,len(result)):
    sen.append(result[i][1])
print(sen)
```

# 2.추출한 텍스트를 번역하기

```python
import gradio as gr
from transformers import pipeline
from transformers import MarianMTModel, MarianTokenizer
```

```python
def translate_transformers(from_text):
    tokenizer=MarianTokenizer.from_pretrained('Helsinki-NLP/opus-mt-ko-en')
    model = MarianMTModel.from_pretrained('Helsinki-NLP/opus-mt-ko-en')
    translated = model.generate(**tokenizer(from_text, return_tensors="pt", padding=True))
    return [tokenizer.decode(t, skip_special_tokens=True) for t in translated]
```

```python
for j in sen:
    print(translate_transformers(j))
```
